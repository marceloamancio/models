#  Linear Regression in Pytorch

* From a set of data $(x_i, y_i)$ find a linear relationship between them.
* We can express this linear relationship as $y = a*x + b + noise$
* Noise can be represented some distribution as the uniform $U(a,b)$ or normal $\mathcal{N}(\mu,\,\sigma^{2})$.

# 1. Generating Data

* Let's generate some data to work with. 

```python
import numpy as np
import torch

def get_data(m_samples, a, b):
    
    # Generate some noise: uniform distribution on interval [0,1]
    random_noise = np.random.normal(0,1,m_samples) # 

    # Generate X input from uniform distribution in range [0, 10] 
    x_input =  np.random.uniform(0,10,m_samples)
    x_input.sort()
    
    #y = a*x + b + noise (univariate linear regression model with noise)
    y_input = a*x_input + b + random_noise
    x_input = torch.from_numpy(x_input.astype(dtype=np.float32).reshape(-1,1))
    y_input = torch.from_numpy(y_input.astype(dtype=np.float32).reshape(-1,1))

    return x_input, y_input

x_train, y_train = get_data(m_samples = 40, a=1.5, b=1)

print('x values')
print(x_train[:3])

print('\n\ny values:')
print(y_train[:3])
```

* **Print Generated Data:**

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

plt.plot(x_train.numpy(), y_train.numpy(), 'ro', label='Linear dependent with noise - Random Generated Data')
plt.xlabel('X VALUES')
plt.xlabel('Y VALUES')
plt.legend() 
plt.show()
```

---

# 2. Pytorch Model

```python
import torch
from torch import nn
from torch.autograd import Variable

class LinearRegressionModel(nn.Module):
    def __init__(self):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(1,1)
        None
    def forward(self,x):
        return self.linear(x)
        
model = LinearRegressionModel()

criterion = torch.nn.MSELoss(size_average = False) 
optimizer = torch.optim.SGD(model.parameters(), lr = 0.00003) 
```

* ** Print Line before optimization **

```python
predict_values = model(Variable(x_train)).data

plt.plot(x_train.numpy(), y_train.numpy(), 'ro', label='Data')
plt.plot(x_train.numpy(), predict_values.numpy(), label='Line before optimization')
plt.legend() 
plt.show()
```

# 3. Optimization Steps

```python
num_epochs = 50
loss_i = []
for epoch in range(num_epochs):
    x = Variable(x_train)
    y = Variable(y_train)

    pred_y = model(x)  #forward
    loss = criterion(pred_y, y)
    
    optimizer.zero_grad() # backward
    loss.backward()
    optimizer.step()
    
    loss_i.append(loss.item())
    
    if (epoch)%10 == 0:
        print('epoch {}, loss {:.6f}'.format(epoch, loss.item()))

print('final epoch {}, loss {:.6f}'.format(num_epochs, loss.item()))  
```

```python
predict_values = model(Variable(x_train)).data

plt.plot(x_train.numpy(), y_train.numpy(), 'ro', label='Data')
plt.plot(x_train.numpy(), predict_values.numpy(), label='Line before optimization')
plt.legend() 
plt.show()
```

---

# 5. Print loss in time i

```python
def print_loss(loss_i):
    x = list(range(len(loss_i)))
    plt.plot(x, loss_i)
    plt.ylabel('time i')
    plt.ylabel('loss value at time i')
    plt.show()
    None
 
print_loss(loss_i)
```
